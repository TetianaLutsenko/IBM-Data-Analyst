```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""

df = pd.read_sql_query(QUERY,conn)
db_counts = df['DatabaseDesireNextYear'].value_counts().head().reset_index()

db_counts.columns = ['DatabaseDesireNextYear', 'count']
db_counts

plt.pie(db_counts['count'], labels=db_counts['DatabaseDesireNextYear'], autopct='%.0f%%')
plt.title('Top 5 Databases desired to learn next year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master where Age<=35 and Age>=30
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)

df_wwh_med = df.groupby(['Age'])['WorkWeekHrs'].median()
df_wwh_med
df_crh_med = df.groupby(['Age'])['CodeRevHrs'].median()
df_crh_med
age = [30, 31, 32, 33, 34, 35]

df_hours = pd.DataFrame({'Age':age, 'WorkWeekHrs': df_wwh_med, 'CodeRevHrs': df_crh_med})
df_hours

df_hours.plot.bar(x='Age')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.xticks(rotation=0)
plt.legend(title='Type of Hours')
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT Age, ConvertedComp
FROM master Where Age>=45 And Age <=60
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df_med = df.groupby(['Age'])['ConvertedComp'].median().reset_index()
df_med
sns.lineplot(df_med, x='Age', y='ConvertedComp')
plt.title('Median ConvertedComp for Age Group 45 to 60')
plt.xlabel('Age')
plt.ylabel('ConvertedComp')
plt.grid()
```
