```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("company_sales_data.csv")
df.head()
```

**Exercise 1:** Read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties:

● X label name = Month Number

● Y label name = Total profit

```python
plt.plot(df['month_number'], df['total_profit'])
plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.title('Total profit')
plt.grid(True)
plt.show()
```

**Exercise 2:** Get total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties:

● Line Style dotted and Line-color should be red

● Show legend at the lower right location.

● X label name = Month Number

● Y label name = Sold units number

● Add a circle marker.

● Line marker color as read

● Line width should be 3

```python
plt.plot(df['month_number'], df['total_profit'], marker='o', linestyle='-.', color='red', linewidth=3)
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit Over Months')
plt.grid(True, linestyle='--', linewidth=0.7)
plt.show()
```

**Exercise 3:** Read all product sales data and show it using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline
for each product).

```python
plt.plot(df['month_number'], df['facecream'], color='red', label = 'facecream')
plt.plot(df['month_number'], df['facewash'], color='blue', label = 'facewash')
plt.plot(df['month_number'], df['toothpaste'], color='green', label = 'toothpaste')
plt.plot(df['month_number'], df['bathingsoap'], color='yellow', label = 'bathingsoap')
plt.plot(df['month_number'], df['shampoo'], color='black', label = 'shampoo')
plt.plot(df['month_number'], df['moisturizer'], color='magenta', label = 'moisturizer')
plt.xlabel('Month Number')
plt.ylabel('Product Sales')
plt.title('Profit for each product over months')
plt.legend()
plt.grid(True, linestyle='--', linewidth=0.7)
plt.show()
```

**Exercise 4:** Read toothpaste sales data of each month and show it using a scatter plot. Also, add a grid in
the plot. gridline style should “–“.

```python
plt.scatter(df['month_number'], df['toothpaste'], color='green', label = 'toothpaste')
plt.xlabel('Month Number')
plt.ylabel('Product Sales')
plt.title('Profit for each product over months')
plt.legend()
plt.grid(True, linestyle='-')
plt.show()
```

**Exercise 5:** Read face cream and facewash product sales data and show it using the bar chart

```python
bar_width = 0.35
# Define the positions of the bars
r1 = np.arange(len(df['month_number']))
r2 = [x + bar_width for x in r1]

plt.bar(r1, df['facecream'], color='blue', width=bar_width, edgecolor='grey', label='Face Cream')
plt.bar(r2, df['facewash'], color='red', width=bar_width, edgecolor='grey', label='Face Wash')

plt.xlabel('Month Number')
plt.ylabel('Product Sales')
plt.title('Profit for Each Product Over Months')
plt.legend()
plt.grid(True, linestyle='--', linewidth=0.7)
plt.show()
```

**Exercise 6:** Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to
your hard disk

```python
plt.bar(df['month_number'], df['bathingsoap'], color='yellow', label = 'bathingsoap')
plt.xlabel('Month Number')
plt.ylabel('Product Sales')
plt.title('Profit for each product over months')
plt.legend()
plt.grid(True, linestyle='--', linewidth=0.7)
plt.savefig('bathingsoap.png')
plt.show()
```

**Exercise 7:** Read the total profit of each month and show it using the histogram to see the most common
profit ranges

```python
num_bins = int(np.sqrt(len(df)))
plt.hist(df['total_profit'], bins=num_bins, edgecolor='black')
plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.title('Distribution of Total Profit')
plt.show()
```

**Exercise 8:** Calculate total sale data for last year for each product and show it using a Pie chart

```python
y = [df['facecream'].sum(), df['facewash'].sum(), df['toothpaste'].sum(), df['bathingsoap'].sum(), df['shampoo'].sum(), df['moisturizer'].sum()]
my_labels = ['Facecreame', 'Facewash', 'Toothpaste', 'Bathingsoap', 'Shampoo', 'Moisturizer']
plt.pie(y, labels = my_labels, startangle = 90, autopct='%1.1f%%')
plt.title('Total profit')
plt.grid(True)
plt.show()
```

**Exercise 9:** Read Bathing soap facewash of all months and display it using the Subplot

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

# The first subplot
ax[0].bar(df['month_number'], df['facewash'], color='orange', label='Facewash')
ax[0].set_xlabel('Month Number')
ax[0].set_ylabel('Sales')
ax[0].set_title('Facewash Sales')
ax[0].legend()
ax[0].grid(True, linestyle='--', linewidth=0.7)

# The second subplot
ax[1].bar(df['month_number'], df['bathingsoap'], color='green', label='Bathingsoap')
ax[1].set_xlabel('Month Number')
ax[1].set_ylabel('Sales')
ax[1].set_title('Bathing Soap Sales')
ax[1].legend()
ax[1].grid(True, linestyle='--', linewidth=0.7)

fig.suptitle('Sales of Bathing Soap and Facewash Over All Months', fontsize=16)

# Adjust layout to prevent overlap
plt.tight_layout(rect=[0, 0, 1, 0.96])

plt.show()
```

**Exercise 10:** Read all product sales data and show it using the stack plot

```python
plt.bar(df['month_number'], df['facecream'], label='Facecream')
plt.bar(df['month_number'], df['facewash'], label='Facewash', bottom=df['facecream'])
plt.bar(df['month_number'], df['toothpaste'], label='Toothpaste', bottom=df['facecream'] + df['facewash'])
plt.bar(df['month_number'], df['bathingsoap'], label='Bathingsoap', bottom=df['facecream'] + df['facewash'] + df['toothpaste'])
plt.bar(df['month_number'], df['shampoo'], label='Shampoo', bottom=df['facecream'] + df['facewash'] + df['toothpaste'] + df['bathingsoap'])
plt.bar(df['month_number'], df['moisturizer'], label='Moisturizer', bottom=df['facecream'] + df['facewash'] + df['toothpaste'] + df['bathingsoap'] + df['shampoo'])

plt.xlabel('Month Number')
plt.ylabel('Product Sales')
plt.title('All Product Sales Data')

plt.legend()
plt.grid(True, linestyle='--', linewidth=0.7)

plt.show()
```
