```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from pywaffle import Waffle
from wordcloud import WordCloud
```

```python
df = pd.read_csv("company_sales_data.csv")
df.head()
```

**Exercise 1:** Show the distribution of total sales among different products. Create waffle charts to display proportions.

```python
product_profits = df[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']].sum()
product_profits.head()

plt.figure(FigureClass= Waffle,
           rows = 10,
           columns = 20,
           values = product_profits,
           title={'label': 'Product Sales Distribution', 'loc': 'center'},
           labels=["{0} ({1})".format(k, v) for k, v in product_profits.items()],
           legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)},
           figsize=(10, 5))
plt.show()
```

**Exercise 2:** Create a word cloud to visualize the popularity of different products based on their total sales.

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(product_profits)

# Plot the word cloud
plt.figure(figsize=(8, 4))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Product Sales Word Cloud')
plt.show()
```

**Exercise 3:** Analyze the trend of total profit over the months.
Create a line plot with a regression line.

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='month_number', y='total_profit', data=df)
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Trend of Total Profit Over Months')
plt.grid(True, linestyle='--', linewidth=0.7)
plt.show()
```

**Exercise 4:** Compare sales of two products

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='month_number', y='facecream', data=df, label='Facecream')
sns.regplot(x='month_number', y='facewash', data=df, label='Facewash')
plt.xlabel('Month Number')
plt.ylabel('Sales')
plt.title('Sales Trends of Facecream and Facewash Over Time')
plt.legend()
plt.grid(True, linestyle='--', linewidth=0.7)
plt.show()
```
